```python
# Longest Collatz Sequence
# For any positive integer:
# if n even, n maps to n/2
# if n odd, n maps to 3n + 1
# This program prints the pos int under 1 million that produces the longest Collatz sequence

longest = 10
longest_num = 13

for i in range (1, 999999):
  n = i
  counter = 1

  while not n == 1:
    counter += 1
    if n % 2 == 0:
      n /= 2
    else:
      n = (3*n) + 1
  
  if counter > longest:
    longest = counter
    longest_num = i

print(longest_num)
```

```python
# This program prints the sum of the digits in 100!

import math as m

num = str(m.factorial(100))
num_length = len(num)
sum = 0

while num_length > 0:
  sum += int(num[0:1])
  num = num[1:num_length]
  num_length -= 1

print(sum)


```

```python
# Given a string containing digits 2-9, return all possible letter combinations
# 2 -> a,b,c
# 3 -> d,e,f
# 4 -> g,h,i
# 5 -> j,k,l
# 6 -> m,n,o
# 7 -> p,q,r,s
# 8 -> t,u,v
# 9 -> w,x,y,z


# returns a new list with the elements of the two given lists concatenated
def concat_lists(a, b):
  c = []
  if len(a) == 0:
    return b
  else:
    for i in range(len(a)):
      for j in range(len(b)):
        c.append(a[i] + b[j])
    return c

# Prints all possible letter combinations of a given phone number
def main():
  possibilities = []
  temp_2d = []
  digits = str(input("Please enter an integer value, not including 1s or 0s: "))

  for i in range(len(digits)):
    curr_digit = digits[i:i+1]
    if curr_digit == "2":
      temp_2d.append(["a", "b", "c"])
    elif curr_digit == "3":
      temp_2d.append(["d", "e", "f"])
    elif curr_digit == "4":
      temp_2d.append(["g", "h", "i"])
    elif curr_digit == "5":
      temp_2d.append(["j", "k", "l"])
    elif curr_digit == "6":
      temp_2d.append(["m", "n", "o"])
    elif curr_digit == "7":
      temp_2d.append(["p", "q", "r", "s"])
    elif curr_digit == "8":
      temp_2d.append(["t", "u", "v"])
    else:
      temp_2d.append(["w", "x", "y", "z"])


  for i in range(len(temp_2d)):
    possibilities = concat_lists(possibilities, temp_2d[i])

  print()
  print("The possible strings for " + digits + " are:")
  print(possibilities)



if __name__ == '__main__':
  main()
```

```python
# Given an integer array height, find two lines that together with the x-axis
# form a container s.t. the container contains the most possible water
# height will always have length at least 2 (so there will always be enough lines
# to form at least one container)


input_array = []
input_array.append(input("Please enter the first height (an integer): "))
input_array.append(input("Please enter the second height (an integer): "))

cont = input("Enter another integer height or enter \"n\" to stop: ")

while not cont == "n":
  input_array.append(cont)
  cont = input("Enter another integer height or enter \"n\" to stop: ")


max_area_line_heights = [int(input_array[0]), int(input_array[1])]
max_area_line_positions = [0, 1]
max_volume = min(max_area_line_heights) * abs(max_area_line_positions[0] - max_area_line_positions[1])

for i in range(len(input_array)-1):
  for j in range(i+1, len(input_array)):
    curr_vol = min(int(input_array[i]), int(input_array[j])) * abs(i - j)
    if curr_vol > max_volume:
      max_volume = curr_vol
      max_area_line_heights = [int(input_array[i]), int(input_array[j])]
      max_area_line_positions = [i, j]

print()
print("The largest volume that can be held is " + str(max_volume) + " between heights " +
      str(max_area_line_heights[0]) + " and " + str(max_area_line_heights[1]))



```
