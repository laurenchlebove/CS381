```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer


df = pd.read_csv("https://raw.githubusercontent.com/MatthewFried/Udemy/master/Day2/Day2_Data.csv")
df_original = df.copy()
print(df.dtypes)
df.head()
```

```python
df.isnull().sum()
```

<h3>Introduction</h3>

There are a los of null values, so I will go through and fill in each value with the median in order to be able to use the data effectively.

```python
# puts median of each column into missing values for that column

for x in df:
  df[x].fillna(df[x].median(), inplace = True)

# delete superfluous index column
df = df.drop("INDEX", axis = 1)

df
```
